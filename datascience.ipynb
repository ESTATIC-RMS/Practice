```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df_cleaned = df.drop(columns=[
    'State Code', 'County Code', 'Site Num', 'Address', 
    'State', 'County', 'City', 'Date Local',
    'NO2 Units', 'O3 Units', 'SO2 Units', 'CO Units'
], errors='ignore')
```

```python
df_cleaned = df_cleaned.dropna()
```

```python
X_mean = df_cleaned.drop(columns=['NO2 Mean', 'NO2 AQI'])
y_mean = df_cleaned['NO2 Mean']
```

```python
X_aqi = df_cleaned.drop(columns=['NO2 Mean', 'NO2 AQI'])
y_aqi = df_cleaned['NO2 AQI']
```

```python
X_mean_train, X_mean_test, y_mean_train, y_mean_test = train_test_split(X_mean, y_mean, test_size=0.2, random_state=42)
X_aqi_train, X_aqi_test, y_aqi_train, y_aqi_test = train_test_split(X_aqi, y_aqi, test_size=0.2, random_state=42)
```

```python
def train_and_evaluate(model, X_train, X_test, y_train, y_test, model_name, target_name):
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    mse = mean_squared_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)
    print(f"\n🔹 {model_name} for {target_name}")
    print(f"➡️ MSE: {mse:.4f}")
    print(f"➡️ R² Score: {r2:.4f}")
```

```python
lr = LinearRegression()
train_and_evaluate(lr, X_mean_train, X_mean_test, y_mean_train, y_mean_test, "Linear Regression", "NO2 Mean")
train_and_evaluate(lr, X_aqi_train, X_aqi_test, y_aqi_train, y_aqi_test, "Linear Regression", "NO2 AQI")
```

```python
dt = DecisionTreeRegressor(random_state=0)
train_and_evaluate(dt, X_mean_train, X_mean_test, y_mean_train, y_mean_test, "Decision Tree", "NO2 Mean")
train_and_evaluate(dt, X_aqi_train, X_aqi_test, y_aqi_train, y_aqi_test, "Decision Tree", "NO2 AQI")
```

```python
rf = RandomForestRegressor(random_state=0)
train_and_evaluate(rf, X_mean_train, X_mean_test, y_mean_train, y_mean_test, "Random Forest", "NO2 Mean")
train_and_evaluate(rf, X_aqi_train, X_aqi_test, y_aqi_train, y_aqi_test, "Random Forest", "NO2 AQI") 
```

```python
gb = GradientBoostingRegressor(random_state=0)
train_and_evaluate(gb, X_mean_train, X_mean_test, y_mean_train, y_mean_test, "Gradient Boosting", "NO2 Mean")
train_and_evaluate(gb, X_aqi_train, X_aqi_test, y_aqi_train, y_aqi_test, "Gradient Boosting", "NO2 AQI")
```
